```python
x=df.drop(columns='y',axis=1)
y=df.iloc[:,[-1]]
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
x['job']=le.fit_transform(x['job'])
x['marital']=le.fit_transform(x['marital'])
x['education']=le.fit_transform(x['education'])
x['default']=le.fit_transform(x['default'])
x['housing']=le.fit_transform(x['housing'])
x['loan']=le.fit_transform(x['loan'])
x['contact']=le.fit_transform(x['contact'])
x['month']=le.fit_transform(x['month'])
x['poutcome']=le.fit_transform(x['poutcome'])
```

```python
x
```

```python
from sklearn.preprocessing import StandardScaler
standardisation=StandardScaler()
res=standardisation.fit_transform(x)
res
```

```python
# train,test X,Y split model selection
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(res,y,test_size=0.3,random_state=42)
```

```python
y_train
```

**KNN Classification**

```python
from sklearn.neighbors import KNeighborsClassifier
classifier= KNeighborsClassifier(n_neighbors=5,p=2 )
classifier.fit(x_train, y_train)
```

```python
y_pred= classifier.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix,recall_score,accuracy_score,precision_score,f1_score
cm = confusion_matrix(y_test,y_pred)
print('Confusion',cm)
accuracy = accuracy_score(y_test,y_pred)
print('Accuracy',accuracy)
#F1score = f1_score(y_test,y_pred)
#print('F1 score',F1score)
```

**support vector machine classification**

```python
from sklearn.svm import SVC
classifier2=SVC(kernel='linear',random_state=0)
classifier2.fit(x_train,y_train)
```

```python
y_pred2= classifier2.predict(x_test)
y_pred2
```

```python
from sklearn.metrics import confusion_matrix,recall_score,accuracy_score,precision_score,f1_score
cm = confusion_matrix(y_test,y_pred2)
print('Confusion',cm)
accuracy = accuracy_score(y_test,y_pred2)
print('Accuracy',accuracy)
```

***Logistic regression classification ***

```python
from sklearn.linear_model import LogisticRegression
Log = LogisticRegression()
Log.fit(x_train,y_train)
```

```python
y_predictlog=Log.predict(x_test)
y_predictlog
```

```python
from sklearn.metrics import confusion_matrix,recall_score,accuracy_score,precision_score,f1_score
cm = confusion_matrix(y_test,y_predictlog)
print('Confusion',cm)
accuracy = accuracy_score(y_test,y_predictlog)
print('Accuracy',accuracy)
```
